```python
import pandas as pd
```

```python
df = pd.read_csv('movies.csv')
display(df)
```

```python
df.info()
df['gross_total'].isnull().sum()
```

```python
df = df.dropna(subset=['gross_total'])
df.info()
```

```python
df = df.rename(columns={'title': 'titulo', 'year': 'ano', 'rating': 'nota IMDB', 'votes': 'numero de votos','runtime': 'duração(min)', 'certificate': 'classificação indicativa', 'genre': 'genero','gross_total': 'bilheteria total(USD)'})
display(df)
```

```python
df['classificação indicativa'].value_counts()
```

```python
mapa_classificacao = {
    'G': 'Livre',
    'PG': '10 anos',
    'PG-13': '13 anos',
    'R': '16 anos',
    'NC-17': '18 anos',
    'Unrated': 'Não classificado'
}
df['classificação indicativa'] = df['classificação indicativa'].map(mapa_classificacao)
display(df)
```

```python
def converter_bilheteria(valor):
    if pd.isna(valor):
        return None
    
    valor = valor.replace('$', '').strip()

    if 'M' in valor:
        return float(valor.replace('M', '')) * 1_000_000
    elif 'K' in valor:
        return float(valor.replace('$', '')) * 1_000
    else:
        return(valor)
    
df['bilheteria total(USD)'] = df['bilheteria total(USD)'].apply(converter_bilheteria)
df.info()
display(df)
```

```python
df['numero de votos'] = df['numero de votos'].str.replace(',', '', regex=False).astype(int)
```

```python
df.sort_values(by='bilheteria total(USD)', ascending=False).head(10)
```

```python
df.to_csv('filmes.csv', index=False)
```
